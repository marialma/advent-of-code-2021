# part 1

```python
file = open("day1_input")
```

```python
lines = file.readlines()
int_list = [int(i) for i in lines]
```

```python
increases = 0
prev_line = 0
```

```python
for line in int_list:
    if prev_line != 0:
        if line > prev_line:
            increases = increases + 1
    prev_line = line
print(increases)
```

# part 2

```python
lag1 = 0 
lag2 = 0
prev_sum = 0
increases = 0
```

```python
for line in int_list: 
    if lag1 == 0 and lag2 == 0:
        lag1 = line
    elif lag1 != 0 and lag2 == 0:
        lag2 = lag1
        lag1 = line
    else: 
        if prev_sum == 0:
            prev_sum = line + lag1 + lag2
            lag2 = lag1
            lag1 = line
        else:
            lagged_sum = line + lag1 + lag2
            lag2 = lag1
            lag1 = line
            if lagged_sum > prev_sum: 
                increases = increases + 1
                prev_sum = lagged_sum
            else:
                prev_sum = lagged_sum
print(increases)
```
